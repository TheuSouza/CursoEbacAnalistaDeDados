```python
import seaborn as sns

data =  sns.load_dataset('diamonds')
data.head()
```

```python
# Outliers do preço do diamante solicitado no exercício.
with sns.axes_style('whitegrid'):
    diamond_outliers = sns.boxplot(x=data['price'])
    diamond_outliers.set(title='Valores discrepantes do peso do diamante', xlabel='Peso do diamante')
    diamond_outliers.figure.set_size_inches(12, 6)
```

__A quantidade de Outliers do preço é grande, provalvelmente isso acontece devido a grande variação do produto, como peso, trasparência e corte.__

```python
diamond_outliers.figure.savefig('diamond_outliers.png', bbox_inches='tight')
```

```python
# Outliers do preço do diamante por tipo de corte.
with sns.axes_style('whitegrid'):
    diamond_outliers_cut = sns.boxplot(x=data['price'], y=data['cut'], hue=data['cut'], palette='Spectral')
    diamond_outliers_cut.set(title='Valores discrepantes do preço do diamante por corte', xlabel='Preço', ylabel='Corte')
    diamond_outliers_cut.figure.set_size_inches(12, 6)
```

```python
diamond_outliers_cut.figure.savefig('diamond_outliers_cut.png', bbox_inches='tight')
```

```python
# Filtrando colunas de interesse.
diamond_clarity_price = data[['carat', 'clarity', 'price']]
diamond_clarity_price.head()
```

```python
# Correlação entre o preço do diamante com seu peso agrupado por transparência.
with sns.axes_style('whitegrid'):
    grafico = sns.scatterplot(data=diamond_clarity_price, x='price', y='carat', hue='clarity', palette='Spectral')
    grafico.set(title='Correlação entre preço e peso do diamante por trasparência', xlabel='Preço', ylabel='Peso')
    grafico.get_legend().set_title('Transparência')
    grafico.figure.set_size_inches(12, 6)
```

__Não sou um especialista em diamantes mas podemos ver que os diamantes com a trasparência l1 geralmente são encontrados em maior quantidade devido o seu peso, consequentemente os mesmos tem o maior valor.__

__O diamante com transparência IF é um diamante com um corte e trasparência melhor, mesmo com peso/tamanho de 2 a 3 vezes menor que o l1, é negociado por um valor muito próximo__

```python
grafico.figure.savefig('diamond_clarity_price.png', bbox_inches='tight')
```

```python
data2 = sns.load_dataset('mpg')
data2.head()
```

```python
# Distribuição contínua aproximada do peso solicitado no exercício.
with sns.axes_style('whitegrid'):
    weight_car = sns.histplot(data=data2, x='weight', kde=True, bins=20)
    weight_car.set(title='Distribuição de peso dos carros', xlabel='Peso', ylabel='Frequência')
    weight_car.figure.set_size_inches(12, 6)
```

__A Distribuição dos carros estão mais concentrada em 2000 libras = 907 kg devido a classe do veículo é o número de cilidros, nessa faixa de peso e com 4 cilindros geralmente são carros populares, por isso a grande quantidade de distribuiçao.__

```python
weight_car.figure.savefig('weight_car.png', bbox_inches='tight')
```

```python
# Distribuição contínua aproximada do peso por número de cilindros.
with sns.axes_style('whitegrid'):
    weight_car_cylinders = sns.histplot(data=data2, x='weight', hue='cylinders', palette='rocket_r', kde=True, bins=20)
    weight_car_cylinders.set(title='Distribuição de peso dos carros por número de cilindros', xlabel='Peso', ylabel='Frequência')
    weight_car_cylinders.get_legend().set_title('Cilindros')
    weight_car_cylinders.figure.set_size_inches(12, 6)
```

```python
weight_car_cylinders.figure.savefig('weight_car_cylinders.png', bbox_inches='tight')
```
