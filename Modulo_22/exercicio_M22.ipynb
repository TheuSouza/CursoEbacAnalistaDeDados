## 1\. Pinguins 

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('penguins')
```

```python
data.head()
```

### **1.1. Valores nulos** 

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```python
data.info()
```

```python
# resposta da questão 1.1

# Separei os valores nulos em um novo dataframe.
data_null = data[data.isnull().any(axis=1)]

# Visualizando.
data_null.head(11)
```

```python
# Descartei os valores Nulos, Os valores númericos também foram descartados pois os mesmos tinhas valores nulos na 
# coluna "sex" e valores categoricos com ensinado devem ser descartados.
data.dropna(inplace=True)
data.isnull().any()
```

### **1.2. Variáveis numéricas** 

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```python
data.head()
```

```python
# resposta da questão 1.2

length_media =  data['bill_length_mm'].mean()
length_desvio_padrao = data['bill_length_mm'].std()

data['bill_length_mm_std'] = data['bill_length_mm'].apply(lambda lenght: (lenght - length_media) / length_desvio_padrao)
```

```python
depth_media = data['bill_depth_mm'].mean()
depth_desvio_padrao = data['bill_depth_mm'].std()

data['bill_depth_mm_std'] = data['bill_depth_mm'].apply(lambda depth: (depth - depth_media) / depth_desvio_padrao)
```

```python
flipper_media = data['flipper_length_mm'].mean()
flipper_desvio_padrao = data['flipper_length_mm'].std()

data['flipper_length_mm_std'] = data['flipper_length_mm'].apply(lambda flipper: (flipper - flipper_media) / flipper_desvio_padrao)
```

```python
body_media = data['body_mass_g'].mean()
body_desvio_padrao = data['body_mass_g'].std()

data['body_mass_g_std'] = data['body_mass_g'].apply(lambda body: (body - body_media) / body_desvio_padrao)

data.head()
```

### **1.3. Variáveis categóricas** 

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```python
data['sex_m_nom'] = data['sex'].apply(lambda sex: 1 if sex == 'Male' else 0)
data['sex_f_nom'] = data['sex'].apply(lambda sex: 1 if sex == 'Female' else 0)
```

```python
data.head()
```

```python
# resposta da questão 1.3

data['species'].drop_duplicates()
```

```python
species_mapper = {
    'Adelie': 1,
    'Chinstrap': 2,
    'Gentoo': 3
}
```

```python
data['species_ord'] = data['species'].apply(lambda specie_ord: species_mapper[specie_ord])
```

```python
data['island'].drop_duplicates()
```

```python
island_mapper = {
    'Torgersen': 1,
    'Biscoe': 2,
    'Dream': 3
}
```

```python
data['island_ord'] = data['island'].apply(lambda island_ord: island_mapper[island_ord])
```
