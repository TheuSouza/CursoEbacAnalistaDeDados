```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.get_dataset_names()
```

```python
diamond = sns.load_dataset('diamonds')
diamond.head()
```

```python
cut_price = (
    diamond[['cut', 'price']]
    .groupby('cut', observed=False)
    .agg(price_sum=('price', 'sum'), price_mean=('price', 'mean'))
    .reset_index()
)

cut_price.head()
```

```python
plt.figure(figsize=(12, 6))
grafico_diamond_mean = sns.barplot(data=cut_price, x='cut', y='price_mean',hue='cut', palette='rocket')
grafico_diamond_mean.set_title('Preço médio do diamante por tipo de corte', fontsize=15)
grafico_diamond_mean.set_xlabel('Corte', fontsize=12)
grafico_diamond_mean.set_ylabel('Preço', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.xticks(rotation=45);
```

```python
grafico_diamond_mean.figure.savefig('price_mean_cut_diamond.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(12, 6))
grafico_diamond_volume = sns.barplot(data=cut_price, x='cut', y='price_sum',hue='cut', palette='rocket')
grafico_diamond_volume.set_title('Valor total negociado do diamante por tipo de corte', fontsize=15)
grafico_diamond_volume.set_xlabel('Corte', fontsize=12)
grafico_diamond_volume.set_ylabel('Preço', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.xticks(rotation=45);
```

```python
grafico_diamond_volume.figure.savefig('value_total_cut_diamond.png', bbox_inches='tight')
```
